# Data Visualization I

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns. load_dataset('titanic')
```

```python
titanic
```

```python
sns.distplot(titanic['fare'])
```

```python
sns.distplot(titanic['fare'], kde = False)
```

```python
sns.distplot(titanic['fare'], hist = False)
```

```python
sns.jointplot(x = 'age', y = 'fare', data = titanic)
```

```python
sns.jointplot(x = 'age', y = 'fare', data = titanic, kind = 'hex')
```

```python
sns.jointplot(x = 'age', y = 'fare', data = titanic, kind = 'hist')
```

```python
plt.plot(titanic['fare'])
```

```python
plt.hist(titanic['fare'])
```

```python
sns.barplot(x="sex",y="age", data = titanic)
```

```python
sns.countplot(x = 'sex', data=titanic)
```
